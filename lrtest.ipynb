```python
from linearregression import LinearRegression
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
X, y = datasets.make_regression(n_samples=100, n_features=1, noise=20, random_state=4)
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
```

```python
maksgay = LinearRegression(0.01)
maksgay.fit(x_train, y_train, 1000)
```

```python
def mse(y, yp):
    return np.mean((y - yp)**2)
```

```python
mse(y_test, maksgay.predict(x_test))
```

```python
import matplotlib.pyplot as plt
```

```python
cmap = plt.get_cmap('viridis')
fig = plt.figure(figsize=(8, 6))
m1 = plt.scatter(x_train, y_train, s=10, color=cmap(0.9))
m2 = plt.scatter(x_test, y_test, s=10, color=cmap(0.5))
plt.plot(X, maksgay.predict(X), color="black")
```
